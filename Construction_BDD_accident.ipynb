```python
import pandas as pd
import os
```

```python
def concat_df_utf(df_name, usecols=None):
    df_return = pd.DataFrame()
    directory = os.fsencode("./Données")

    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.startswith(df_name): 
            if filename.endswith("2021.csv") or filename.endswith("2020.csv") or filename.endswith("2019.csv"): 
                #print(filename)
                df = pd.read_csv(os.path.join("./Données", filename), 
                                 on_bad_lines='skip', sep = ";", usecols = usecols)
                #print(df.columns)
            else:
                df = pd.read_csv(os.path.join("./Données", filename), on_bad_lines='skip', usecols = usecols)


            df_return = pd.concat([df_return, df], ignore_index=True, join='outer')
    return df_return

             # print(os.path.join(directory, filename))

def concat_df_iso(df_name, usecols=None):
    df_return = pd.DataFrame()
    directory = os.fsencode("./Données")

    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.startswith(df_name): 

            if filename.endswith("2021.csv") or filename.endswith("2020.csv") or filename.endswith("2019.csv"): 
                df = pd.read_csv(os.path.join("./Données", filename), on_bad_lines='skip', sep = ";",
                                 encoding = "ISO-8859-1", usecols = usecols)
            elif filename.endswith("2009.csv"):
                df = pd.read_csv(os.path.join("./Données", filename), 
                                 on_bad_lines='skip', sep = "\t", encoding = "ISO-8859-1", usecols = usecols)
            else:
                df = pd.read_csv(os.path.join("./Données", filename), on_bad_lines='skip', 
                                 encoding = "ISO-8859-1", usecols = usecols)


            df_return = pd.concat([df_return, df], ignore_index=True, join='outer')
    return df_return

             # print(os.path.join(directory, filename))
```

```python
col = ['Num_Acc', 'catr', 'voie', 'v1', 'v2', 'circ', 'nbv',
       'vosp', 'prof', 'plan', 'larrout', 'surf', 'infra', 'situ']
lieux = concat_df_utf("lieux", usecols = col)
```

```python
col = ['Num_Acc', 'an', 'mois', 'jour', 'hrmn', 'agg', 'adr', 'lat', 'long', "dep"]
caracteristiques = concat_df_iso("caracteristiques", usecols = col)
```

```python
col = ['Num_Acc', 'grav']
usagers = concat_df_utf("usagers", usecols = col)
usagers.columns
```

```python
col = ['Num_Acc', 'catv']
vehicules = concat_df_utf("vehicules", usecols = col)
vehicules.columns
```

```python
from functools import reduce
df = reduce(lambda df1,df2: pd.merge(df1,df2,on='Num_Acc'), [caracteristiques, vehicules, usagers, lieux])
```

```python
df["dep"] = df["dep"].replace(750,"75")
```

```python
df.groupby("dep").count().sort_values(by="Num_Acc")
```

```python
df_paris = df[df["dep"]=="75"]
```
