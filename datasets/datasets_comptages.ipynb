```python
import pandas as pd
from os import path
```

```python
data_folder = "C:\\Users\\robin\\Documents\\CS\\Cours\\IA\\Challenge étude de cas\\data"
comptage = path.join(data_folder, "Comptage")
```

```python
velos_paris = path.join(comptage, "comptages-velos-paris.csv")
velos_toulouse = path.join(comptage, "comptages-velos-toulouse.csv")
velos_rennes = path.join(comptage, "comptages-velos-rennes.csv")
```

```python
with open(velos_paris, 'r') as paris:
    df_paris = pd.read_csv(paris, delimiter=";", usecols=['id_compteur', 'nom_compteur', 'id', 'name', 'sum_counts', 'date',
       'installation_date', 'coordinates', 'mois_annee_comptage'])
df_paris.head()
```

```python
df_paris.columns
```

```python
df_paris['isoday'] = df_paris['date'].apply(lambda x : x[:10])
daily_sum_paris = df_paris.groupby('isoday').agg({
    'name': ['nunique'],
    'sum_counts': ['sum']
})
daily_sum_paris.head(20)
```

```python
daily_sum_paris.to_csv('comptages_journaliers_paris.csv')
```

```python
with open(velos_toulouse, 'r') as toulouse: 
    df_toulouse = pd.read_csv(toulouse, delimiter=";")
df_toulouse.head()
```

```python
df_toulouse.columns
```

```python
with open(velos_rennes, 'r') as rennes: 
    df_rennes = pd.read_csv(rennes, delimiter=";")
df_rennes.head()
```

```python
df_rennes['isoday'] = df_rennes['isodate'].apply(lambda x : x[:10])
df_rennes.head()
```

```python
df_rennes_grouped = df_rennes.groupby('isoday').agg({
    'name': ['nunique'],
    'counts': ['sum']
})
df_rennes_grouped.head(3000)
```

```python
df_rennes_grouped.to_csv('comptages_journaliers_rennes.csv')
```
