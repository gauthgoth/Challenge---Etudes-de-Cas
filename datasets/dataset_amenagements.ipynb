```python
# Imports

import pandas as pd
import geopandas as gpd
from os.path import join

from OSMPythonTools.api import Api
from OSMPythonTools.overpass import Overpass
```

```python
# Load data

data_folder = "C:\\Users\\robin\\Documents\\CS\\Cours\\IA\\Challenge étude de cas\\data"
infra_folder = join(data_folder, "Aménagements")

#infra = join(infra_folder, 'test_amenagement.geojson') #FIXME : temporarily lighter dataset loaded to save memory
infra = join(infra_folder, 'france-20230101.geojson')

with open(infra, 'r') as infra_file:
    infra_df = gpd.read_file(infra_file)

infra_df.head()
```

```python
# Get valid INSEE codes for communes within Paris or Rennes metropoles

insee_folder = join(data_folder, "Communes")


insee_rennes = join(insee_folder, 'communes-rennes.csv')
insee_paris = join(insee_folder, 'communes-paris.csv')

with open(insee_rennes, 'r') as rennes:
    insee_rennes_df = pd.read_csv(rennes, delimiter=';', usecols=['geolocalisation', 'code_insee', 'commune'])

with open(insee_paris, 'r') as paris: 
    insee_paris_df = pd.read_csv(paris, delimiter=';', on_bad_lines='skip', usecols=['Geo Shape', 'insee', 'nomcom'])
```
